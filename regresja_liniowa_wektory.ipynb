```python
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
```

dane:

(8,3)

```python
x = np.array([2,5,7,8]).reshape((4,1))

X = np.array([
    [1,2],
    [1,5],
    [1,7],
    [1,8],
])

y = np.array([1,2,3,3]).reshape((4,1))
```

```python
X
```

```python
len(y)
```

```python
y
```

```python
x
```

```python
B = inv(np.transpose(X) @ X) @ np.transpose(X) @ y
B
```

```python
B_0 = B[0][0]
B_1 = B[1][0]
```

B_0 + B_1 * X = y

```python
B_0  # theta_0
```

```python
B_1  # theta_1
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, alpha=1)
plt.plot(x, [(q * B_1) + B_0 for q in x], color='red')
plt.title('Regresja liniowa')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
